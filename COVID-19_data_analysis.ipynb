```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('owid-covid-data.csv')
df.head()
```

```python
india=df[df['location']=='India']
india=india[['date','total_cases','new_cases','total_deaths','new_deaths','people_vaccinated']]
india['date']=pd.to_datetime(india['date'])
india.set_index('date',inplace=True)
india.tail()
```

```python
#total cases over time
plt.figure(figsize=(10,4))
plt.plot(india['total_cases'],label='Total Cases',color='green')
plt.title('Total COVID-19 Cases in India over time')
plt.xlabel('Date')
plt.ylabel('Cases')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#New cases Trend
plt.figure(figsize=(10,4))
plt.plot(india['new_cases'],label='New Cases',color='blue')
plt.title('Daily New COVID-19 Cases in India')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Vaccination Progress
plt.figure(figsize=(10,4))
plt.plot(india['people_vaccinated'],label='People Vaccinated',color='orange')
plt.title('COVID-19 Vaccination Progress in India')
plt.xlabel('Date')
plt.ylabel('People Vaccinated')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Summary Stats
print('Peak Daily Cases: ',india['new_cases'].max())
print('Total Deaths: ',india['total_deaths'].max())
print('People Vaccinated: ',india['people_vaccinated'].max())
```

```python
#India vs USA COVID-19 Trends Comparison
countries=df[df['location'].isin(['India','United States'])]
countries=countries[['location','date','total_cases','new_cases']]
countries['date']=pd.to_datetime(countries['date'])
plt.figure(figsize=(12,6))
sns.lineplot(data=countries,x='date',y='total_cases',hue='location')
plt.title('Total COVID-19 Cases: India vs USA')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=countries,x='date',y='new_cases',hue='location')
plt.title('Daily New COVID-19 Cases: India vs USA')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.grid(True)
plt.show()
```

```python
#7-day rolling average for new cases
india['new_cases_rolling']=india['new_cases'].rolling(window=7).mean()
plt.figure(figsize=(12,5))
plt.plot(india.index,india['new_cases'],alpha=0.4,label='Daily New Cases')
plt.plot(india.index,india['new_cases_rolling'],color='red',label='7-Day Rolling Average')
plt.title('Daily New COVID-19 Cases in India with Rolling Average')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#correlation Heatmap
plt.figure(figsize=(8,5))
correlation=india.corr(numeric_only=True)
sns.heatmap(correlation,annot=True,cmap='coolwarm')
plt.title('Correlation Matrix-COVID Features(India)')
plt.show()
```

```python
#top 10 countries by vaccination or death rate
latest=df[df['date']==df['date'].max()]
top_vaccinated=latest[['location','people_vaccinated']].dropna().sort_values(by='people_vaccinated',ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x='people_vaccinated',y='location',data=top_vaccinated,palette='Greens_r')
plt.title('Top 10 Countries by People Vaccinated')
plt.xlabel('People Vaccinated')
plt.ylabel('Country')
plt.show()
```
